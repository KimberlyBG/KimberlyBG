```python
# O programa irar procurar a existencia do mesmo numero de Telefone
# se encontrar acusarar que ja existe um contato registrado.
def existe_contato(lista, telefone):
    if len(lista) > 0:
        for contato in lista:
            if contato["telefone"] == telefone:
                return True
    return False

# O usuario irar cadatras o novo telefone em sua agenda.
def adicionar(lista):
    while True:
        telefone = int(input("Digite o Telefone de contato: "))
        if not existe_contato(lista, telefone):
            break
        else:
            print("Telefone Ja ultilizado!")
            print("Insira outro Telefone para cadastro.")

    contato = {
        "nome": str(input("Digite o nome: ")),
        "telefone": telefone,
        "email": str(input("Digite o e-mail do contato:")),
        "twitter": str(input("Digite o twitter: ")),
        "instagram": str(input("Digite o instagram: "))
    }

    lista.append(contato)
    print("O contato {} foi cadastrado com sucesso!\n".format(contato["nome"]))


# o usurio poderar salvar sua agenda em um documento
def salvar_contatos(lista):
    arquivo = open("contatos.txt", "w")

    for contatos in lista:
        arquivo.write("{},{},{},{},{}\n".format(
            contatos["nome"], contatos["telefone"], contatos["email"], contatos["twitter"], contatos["instagram"]))
    arquivo.close()


# o usuario opoderar listar a quantidade de contatos em sua lista
def carregar_contatos(lista):
    try:
        arquivo = open("contatos.txt", "r")

        for linha in arquivo.readline():
            coluna = linha.strip().split(",")

            contato = {
                "nome": coluna[0],
                "telefone": coluna[1],
                "email": coluna[2],
                "twitter": coluna[3],
                "instagram": coluna[4]
            }

            lista.append(contato)

        arquivo.close()

        return lista
    except FileNotFoundError:
        pass


# o usurio poderar alterar o cadastro do registro da agenda
def alterar(lista):
    print("\n======= Alterar Contato =========")
    if len(lista) > 0:
        telefone = input("Digite o telefone do contato a ser alterado: ")
        if existe_contato(lista, telefone):
            for contato in lista:
                if contato["telefone"] == telefone:
                    print("\n")
                    print(".............................................")
                    print("Nome: {}".format(contato["nome"]))
                    print("Telefone: {}".format(contato["telefone"]))
                    print("Email: {}".format(contato["e-mail"]))
                    print("Twitter: {}".format(contato["twitter"]))
                    print("Instagram: {}".format(contato["instagram"]))
                    print("..............................................")
                    print("\n")

                    contato["nome"] = input("Digite o novo nome do contato: ")
                    contato["telefone"] = input(
                        "Digite o novo telefone do contato: ")
                    contato["e-email"] = input(
                        "Digite o novo e-mail do contato: ")
                    contato["twitter"] = input(
                        "Digite o novo twitter do contato: ")
                    contato["instagram"] = input(
                        "Digite o novo instagram do contato: ")

                    print("Os dados do contato com o telefone {}, foram alterados com sucesso!\n".format(
                        contato["telefone"]))
                    break


# O usuario deve buscar pelo telefone que deseja excluir , excluindo assim todo o contato.
def excluir(lista):
    print("\n Excluir contato:")
    if len(lista) > 0:
        telefone = int(input("Digite o telefone do contato a ser excluído:..."))
        if existe_contato(lista, telefone):
            for i, contato in enumerate(lista):
                if contato["telefone"] == telefone:
                    print("\n==============================================")
                    print("Nome: {}".format(contato["nome"]))
                    print("Telefone: {}".format(contato["telefone"]))
                    print("Email: {}".format(contato["email"]))
                    print("Twitter: {}".format(contato["twitter"]))
                    print("Instagram: {}".format(contato["instagram"]))
                    print("==============================================\n")

                    del lista[i]

                    print("o contato foi apagado com sucesso")
                    break

        else:
            print(
                "Não existe contato cadastrado no sistema com o email {}.\n".format(telefone))
    else:
        print("Não existe nenhum contato cadastrado no sistema.\n")


# O Usurio deve digitar o telefone que deseja buscar ,assim ele irar procurar dos telefones ja existetes na lista.
def buscar(lista):
    print("\nBuscar Contato ......")
    if len(lista) > 0:
        telefone = input("Digite o Telefone do contato a ser encontrado: ")
        if existe_contato(lista, telefone):
            for contato in lista:
                if contato["telefone"] == telefone:
                    print("\n")
                    print("...........................................")
                    print("Nome: {}".format(contato["nome"]))
                    print("Telefone: {}".format(contato["telefone"]))
                    print("Email: {}".format(contato["email"]))
                    print("Twitter: {}".format(contato["twitter"]))
                    print("Instagram: {}".format(contato["instagram"]))
                    print(".............................................")
                    print("\n")
                    break
        else:
            print(
                "Não existe contato cadastrado no sistema com esse telefone {}.\n".format(telefone))
    else:
        print("Não existe nenhum contato cadastrado no sistema.\n")


# Lista de cada contado registrado na base do programa.
def listar(lista):
    if len(lista) > 0:
        for i, contato in enumerate(lista):
            print("\n")
            print("==============================================")
            print("contato {}:".format(i + 1))
            print("\tNome: {}".format(contato["nome"]))
            print("\tTelefone: {}".format(contato["telefone"]))
            print("\tEmail: {}".format(contato["email"]))
            print("\tTwitter: {}".format(contato["twitter"]))
            print("\tInstagram: {}".format(contato["instagram"]))
            print("==============================================")
            print("")

        print("Quantidade de contatos: {}\n".format(len(lista)))

    else:
        print("Não existe nenhum contato cadastrado para listar.\n")


# MenuP (Menu Principal onde o Usuario Vai selecionbar a opcao que deseja!
def MenuP():
    lista = []
    carregar_contatos(lista)

    while True:
        print("................................")
        print("...... AGENDA DE CONTATOS ......")
        print("................................")
        print("...~> 1- Adicionar contato <~...")
        print("...~> 2-  Alterar contato  <~...")
        print("...~> 3-  Excluir contato  <~...")
        print("...~> 4-  Buscar contato   <~...")
        print("...~> 5-  Lista contato    <~...")
        print("...~> 6-  SAIR             <~...")
        print("..............................")
        while True:
            try:
                opc = int(input("\n Escolha uma das opões acima ~> "))
                opc = int(opc)
                break
            except ValueError:
                print("Nao encontrei um numero !\n Por favor digite novamente! ")

        if opc == 1:
            adicionar(lista)
            salvar_contatos(lista)
        elif opc == 2:
            alterar(lista)
            salvar_contatos(lista)
        elif opc == 3:
            excluir(lista)
            salvar_contatos(lista)
        elif opc == 4:
            buscar(lista)
        elif opc == 5:
            listar(lista)
        elif opc == 6:
            print("\nSaindo do Programa...\nExecução finalizada com sucesso !\n")
            break
        else:
            print("\nOpção inválida. Por favor, tente novamente\n")


MenuP()
```
